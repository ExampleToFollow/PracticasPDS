# IEE352 - Downsampling
Es el proceso mediante el cual se reducen las muestras de una señal discreta en un factor $M$. También se le conoce como compresor.
$$
\overset{x[n]}{\longrightarrow}
{\boxed{\:\downarrow M\:}}
\overset{x_M[n]}{\longrightarrow}
$$

$$x_M[n] = x[nM]=\sum_{k=-\infty}^{\infty} x[k] \delta[nM-k].$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 4000
B = 500
Ns = 4000
n = np.arange(Ns)
t = n/fs
chirp = (1-t)*np.sin(2*np.pi*(B/2)*t**2)
```

```python
plt.figure(1)
plt.plot(t,chirp)
plt.xlim([0,0.5])
plt.grid('on')
plt.xlabel('Time [sec]')
plt.ylabel('Chirp signal')
```

```python
fft_chirp = np.fft.fftshift(np.fft.fft(chirp))
spc = np.abs(fft_chirp)**2
freq = (np.arange(Ns) - Ns/2) * fs / Ns
```

```python
plt.figure(2)
plt.plot(freq,spc,label=str(B)+' Hz/sec')
plt.xlim([-fs/2,fs/2])
plt.grid('on')
plt.xlabel('Frequency [Hz]')
plt.legend(loc='upper left')
```

```python
D = 4
Ns_dn = Ns/D
fs_dn = fs/D
n_dn = np.arange(Ns_dn)
chirp_dn = chirp[::D]
t_dn = n_dn/fs_dn
```

```python
plt.figure(3)
plt.plot(t_dn,chirp_dn,'o')
plt.xlim([0,0.5])
plt.xlim([0.5,0.51])
plt.grid('on')
plt.xlabel('Time [sec]')
plt.ylabel('Chirp signal Upsampled')
```

```python
fft_chirp_dn = np.fft.fftshift(np.fft.fft(chirp_dn))
spc_dn = np.abs(fft_chirp_dn)**2
freq_dn = (n_dn - Ns_dn/2) * fs_dn / Ns_dn
```

```python
plt.figure(3)
plt.plot(freq_dn,spc_dn,label='Chirp')
plt.xlim([-fs_dn/2,fs_dn/2])
plt.grid('on')
plt.xlabel('Frequency [Hz]')
plt.legend(loc='upper left')
plt.show()
```
