# 1 <a id="section-1"></a>  Numpy

## 1.1 <a id="subsection-1.1"></a> Importando librerias

```python
import numpy as np
print("NumPy importado con éxito.")
```

## 1.2 <a id="subsection-1.2"></a> Creando arrays

```python
a = np.array([1, 2, 3])
print("Array a: ", a)

b = np.array([[1, 2], [3, 4]])
print("Array b: \n", b)

c = np.array([1, 2, 3])
print("Array a: ", a)

b = np.array([1, 2, 3])
```

## 1.3 <a id="subsection-1.3"></a> Indexación

```python
print("Primer elemento de a: ", a[0])
print("Rango de elementos de a: ", a[1:3])
```

## 1.4 <a id="subsection-1.4"></a>  Operaciones con arrays

```python
c = a + b
print("Suma de arrays a y a: \n", c)

d = 2 * a
print("Multiplicación de a por 2: ", d)

e = np.mean(a)
print("Media de los elementos de a: ", e)

max_val = np.max(a)
print("Valor máximo de a: ", max_val)

min_val = np.min(a)
print("Valor mínimo de a: ", min_val)

sum_val = np.sum(a)
print("Suma de los elementos de a: ", sum_val)
```

```python
c = np.multiply(a, b)
print("Multiplicación elemento a elemento de a y b: \n", c)

d = np.dot(a, b)
print("Producto punto de a y b: \n", d)

e = np.cross(a, b)
print("Producto cruz de a y b: \n", e)

f = a @ b
print("Multiplicación matricial de a y b: \n", f)
```

```python
# np.arange()
arr1 = np.arange(0, 10, 2)
print("Vector de 0 a 10-1 de 2 en 2 ", arr1)

# np.zeros()
arr2 = np.zeros(5)
print("Vector de ceros con 5 elementos", arr2)
# np.linspace()
arr3 = np.linspace(0, 10, num=5)
print("Vector de 5 elementos igualmente espaciados de 0 a 10", arr3)
```

## 1.5 <a id="subsection-1.5"></a> Normas de vectores

```python
v = np.array([1, 2, 3, 4, 5])
norm1 = np.linalg.norm(v, ord=1)
print("Norma 1 de v: ", norm1)

norm2 = np.linalg.norm(v, ord=2)
print("Norma 2 de v: ", norm2)

norm_inf = np.linalg.norm(v, ord=np.inf)
print("Norma infinita de v: ", norm_inf)
```

# 2 <a id="section-2"></a> Matplotlib

## 2.1 <a id="subsection-2.1"></a> Importando Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2.2 <a id="subsection-2.2"></a> Ejemplo de gráfico de línea

```python
x = np.linspace(0, 2*np.pi*5, 1000)
y = np.sin(x)
plt.plot(x, y)
plt.title("Gráfico de línea")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

## 2.3 <a id="subsection-2.3"></a> Cambiar tamaño de fuente

```python
plt.plot(x, y, label="Mi gráfico")
plt.title("Título", fontsize=20)
plt.legend(fontsize=14)
plt.legend(loc='lower right', fontsize= 14)
plt.xlabel("Eje x", fontsize=16)
plt.ylabel("Eje y", fontsize=16)
```

## 2.4 <a id="subsection-2.4"></a> Cambiar colores

```python
plt.plot(x, y, color="r")
```

```python
plt.plot(x, y, color="g")
```

```python
plt.plot(x, y, color="b")
plt.scatter(x, y, c=x, cmap="coolwarm")
```

## 2.5 <a id="subsection-2.5"></a> Tipos de marker

```python
x = np.linspace(0, 2*np.pi*1.5, 100)
y = np.sin(x)

plt.plot(x, y, marker="o", markersize=5, markeredgecolor="b")
```

```python
plt.plot(x, y, marker="s", markersize=5, markeredgecolor="b")
```

```python
plt.plot(x, y, marker="*", markersize=7, markeredgecolor="b")
```

## 2.6 <a id="subsection-2.6"></a> Tipos de graficos

## Ejemplo de gráfico de barras

```python
x = ["A", "B", "C", "D", "E"]
y = [3, 6, 2, 8, 4]
plt.bar(x, y)
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

## Ejemplo de gráfico de dispersión

```python

x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.title("Gráfico de dispersión")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

## Ejemplo de gráfico de áreas

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, y2, alpha=0.5)
plt.title("Gráfico de áreas")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

## 2.7 <a id="subsection-2.7"></a> Multiples gráficos en una imagen

```python
x = np.linspace(0, 2*np.pi*1.5, 100)
y = np.sin(x)

## Comando adicional para cambiar el tamano de los graficos en el notebook
plt.figure(figsize=(18,10))
plt.plot(x, 2*y, '-o', label="Mi gráfico 1", linewidth=2)
plt.plot(x, y, '-o', label="Mi gráfico 2", linewidth=2)
plt.plot(x, y*0.5, '-o', label="Mi gráfico 3", linewidth=2)

plt.title("Graficos Multiples", fontsize=30)
plt.legend(fontsize=24)
plt.xlabel("Eje x", fontsize=26)
plt.ylabel("Eje y", fontsize=26)
plt.show()
```
