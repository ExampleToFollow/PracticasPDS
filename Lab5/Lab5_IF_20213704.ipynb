Pregunta 3 :
Sea X una variable aleatoria que representa el resultado de un dado trucado y que tiene las
siguientes probabilidades:
P(X=1) = 0.30
P(X=2) = 0.20
P(X=3) = 0.15
P(X=4) = 0.20
P(X=5) = 0.05
P(X=6) = 0.10

a) Simule 1000 veces el lanzamiento del dado trucado mencionado a partir de un generador
de n√∫meros aleatorios uniformemente distribuidos entre 0 y 1. Grafique el histograma de
los resultados y compare con la distribuci√≥n te√≥rica (graficarla). ¬øSon similares los
resultados? ¬øQu√© deber√≠a realizarse para que los resultados se aproximen m√°s a la
distribuci√≥n te√≥rica y por qu√©? (1.0 pts.)


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las probabilidades y los posibles resultados del dado
valores = [1, 2, 3, 4, 5, 6]
probabilidades = [0.30, 0.20, 0.15, 0.20, 0.05, 0.10]

# Simulaci√≥n del inciso a)
# Simular 1000 lanzamientos de un dado truncado
np.random.seed(0)  # Para reproducibilidad
lanzamientos = np.random.choice(valores, size=1000, p=probabilidades)

# Graficar el histograma de los resultados simulados
plt.figure(figsize=(10, 5))
plt.hist(lanzamientos, bins=np.arange(0.5, 7.5, 1), density=True, rwidth=0.8, color='skyblue', edgecolor='black', align='mid')
plt.xlabel('Valor del Dado')
plt.ylabel('Frecuencia Relativa')
plt.title('Histograma de 1000 Lanzamientos de un Dado Truncado')

# Graficar la distribuci√≥n te√≥rica sobre el histograma
plt.plot(valores, probabilidades, 'ro-', label='Distribuci√≥n Te√≥rica')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Comentario sobre la aproximaci√≥n a la distribuci√≥n te√≥rica
print("Para que los resultados se aproximen m√°s a la distribuci√≥n te√≥rica, ser√≠a necesario incrementar el n√∫mero de simulaciones. Seg√∫n la Ley de los Grandes N√∫meros, a medida que aumentamos el n√∫mero de lanzamientos, la distribuci√≥n emp√≠rica se aproximar√° a la distribuci√≥n te√≥rica.")

# Simulaci√≥n del inciso b)
# Simular 1000 veces la suma de 5 lanzamientos de dados truncados
suma_lanzamientos = [np.sum(np.random.choice(valores, size=5, p=probabilidades)) for _ in range(1000)]

# Graficar el histograma de las sumas de los lanzamientos
plt.figure(figsize=(10, 5))
plt.hist(suma_lanzamientos, bins=range(5, 31), density=True, rwidth=0.8, color='lightgreen', edgecolor='black', align='mid')
plt.xlabel('Suma de 5 Lanzamientos')
plt.ylabel('Frecuencia Relativa')
plt.title('Histograma de la Suma de 5 Lanzamientos de un Dado Truncado')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Explicaci√≥n sobre el Teorema del L√≠mite Central
print("Seg√∫n el Teorema del L√≠mite Central, la suma de variables aleatorias independientes tiende a distribuirse de manera aproximadamente normal a medida que aumenta el n√∫mero de variables sumadas. Por lo tanto, en este caso, la distribuci√≥n de las sumas deber√≠a aproximarse a una distribuci√≥n normal.")
```
