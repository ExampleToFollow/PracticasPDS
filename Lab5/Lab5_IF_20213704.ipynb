c) Grafique la densidad espectral de potencia de las señales s1 y s2. (1.0 pts.)


```python
def fourier(signal):
    espectro = np.fft.fftshift(np.fft.fft(signal))
    freq = 2*np.pi*(np.arange(len(signal))-len(signal)/2)/len(signal)
    return freq , np.abs(espectro) , np.angle(espectro)


# Calcular los espectros
freq1, magnitud1, fase1 = fourier(s1)
freq2, magnitud2, fase2 = fourier(s2)

# Crear las gráficas
plt.figure(figsize=(14, 10))

# Espectro de Magnitud para s1
plt.subplot(2, 2, 1)
plt.plot(freq1, magnitud1, linewidth=1.5)
plt.title("Espectro de Magnitud de s1", fontsize=14)
plt.xlabel("Frecuencia [Hz]", fontsize=12)
plt.ylabel("Magnitud", fontsize=12)
plt.grid(True)

# Espectro de Fase para s1
plt.subplot(2, 2, 2)
plt.plot(freq1, fase1, linewidth=1.5, color='orange')
plt.title("Espectro de Fase de s1", fontsize=14)
plt.xlabel("Frecuencia [Hz]", fontsize=12)
plt.ylabel("Fase [radianes]", fontsize=12)
plt.grid(True)

# Espectro de Magnitud para s2
plt.subplot(2, 2, 3)
plt.plot(freq2, magnitud2, linewidth=1.5)
plt.title("Espectro de Magnitud de s2", fontsize=14)
plt.xlabel("Frecuencia [Hz]", fontsize=12)
plt.ylabel("Magnitud", fontsize=12)
plt.grid(True)

# Espectro de Fase para s2
plt.subplot(2, 2, 4)
plt.plot(freq2, fase2, linewidth=1.5, color='orange')
plt.title("Espectro de Fase de s2", fontsize=14)
plt.xlabel("Frecuencia [Hz]", fontsize=12)
plt.ylabel("Fase [radianes]", fontsize=12)
plt.grid(True)

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

e) Considere el uso de un filtro, tal que
𝑦[𝑛] = 0.4𝑦[𝑛 − 1] + 0.4𝑦[𝑛 − 2] + 𝑥[𝑛]
Aplique el filtro a la señal identificada como Ruido Blanco Gaussiano en la pregunta (a)
para obtener y[n]. Así, halle la varianza de y[n] de modo experimental y grafique su
autocorrelación. ¿Cómo afecta el filtro a la señal original? (1.0 pts.)


```python
#La señal identifica como ruido blanco fue la s1
from scipy.signal import lfilter

# Definir el filtro y[n] = 0.4y[n-1] + 0.4y[n-2] + x[n]
b = [1]  # Coeficientes del numerador (x[n])
a = [1, -0.4, -0.4]  # Coeficientes del denominador (y[n-1], y[n-2])

# Aplicar el filtro a la señal de ruido blanco
y = lfilter(b,a, s1)

# Calcular la varianza de la señal filtrada
varianza_y = np.var(y)
print(f"Varianza de la señal filtrada: {varianza_y}")
varianza_s1 = np.var(s1)
print(f"Varianza de la señal original: {varianza_s1}")
# Función para calcular la autocorrelación
def autocorrelacion(signal):
    return  np.correlate(signal, signal, mode='full')/ len(signal)

# Calcular la autocorrelación de las señales
autocorr_x = autocorrelacion(s1)
autocorr_y = autocorrelacion(y)

# Graficar la autocorrelación de la señal original y filtrada
plt.figure(figsize=(14, 6))
index= np.arange( -len(autocorr_x)/2 , len(autocorr_x)/2 )
# Autocorrelación de la señal original
plt.subplot(1, 2, 1)
plt.plot(index , autocorr_x, label="Autocorrelación de s1 (ruido blanco)")
plt.title("Autocorrelación de la Señal Original s1", fontsize=14)
plt.xlabel("Indice", fontsize=12)
plt.ylabel("Autocorrelación", fontsize=12)
plt.grid(True)
plt.legend()

# Autocorrelación de la señal filtrada
plt.subplot(1, 2, 2)
plt.plot(index , autocorr_y, label="Autocorrelación de y (señal filtrada)", color='orange')
plt.title("Autocorrelación de la Señal Filtrada (y)", fontsize=14)
plt.xlabel("Indice", fontsize=12)
plt.ylabel("Autocorrelación", fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scipy.signal import firwin, freqz
dirac = np.zeros(1000)
dirac[0] = 1
w , H_z = freqz(lfilter(b,a, dirac))
H_z  = H_z ** 2
plt.title("Respuesta en frecuencia del filtro al cuadrado")
plt.plot(w,H_z)
```

El filtro aplicado a la señal de entrada en realidad es un LowPassFilter, este filtor elimina las componentes de frecuencias altas por lo que se ve reflejado en la salida de la autocorrelación del filtro ya que no existen cambios tan rapidos principalmente en el centro ya no es un pico tipo delta de dirac como se debería de representar idealmente sino se volvio más n pico triangular representando la eliminación de las componentes de alta frecuencia que serían q se traducen en cambios rápidos en la señal.

a) Considerando el ejemplo, generar una variable aleatoria uniforme 𝑋 con 10000 muestras

densidad de probabilidad (histograma) de las variables aleatorias 𝑋 e 𝑌. Considere una
resolución para los histogramas igual a 50 bins. (1.0 pts.)


```python
import numpy as np
import matplotlib.pyplot as plt

# Número de muestras
n_muestras = 10000

# Generar X ~ U(0,1)
X = np.random.uniform(0, 1, n_muestras)

# Transformación inversa: g(x) = x^(1/3)
Y = X**(1/3)

# Configuración de los histogramas
bins = 50

# Crear gráficos
plt.figure(figsize=(12, 6))

# Histograma de X
plt.subplot(1, 2, 1)
plt.hist(X, bins=bins, density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title("Histograma de X ~ U(0,1)", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.grid(True)

# Histograma de Y
plt.subplot(1, 2, 2)
plt.hist(Y, bins=bins, density=True, alpha=0.75, color='green', edgecolor='black')
plt.title("Histograma de Y ~ f(y) = 3y^2", fontsize=14)
plt.xlabel("Y", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las probabilidades y los posibles resultados del dado
valores = [1, 2, 3, 4, 5, 6]
probabilidades = [0.30, 0.20, 0.15, 0.20, 0.05, 0.10]

# Simular 1000 lanzamientos de un dado truncado
np.random.seed(20213704)  # Para reproducibilidad
lanzamientos = np.random.choice(valores, size=1000, p=probabilidades)

# Graficar el histograma de los resultados simulados
plt.figure(figsize=(10, 5))
plt.hist(lanzamientos, bins=np.arange(0.5, 7.5, 1), density=True, rwidth=0.8, color='skyblue', edgecolor='black', align='mid')
plt.xlabel('Valor del Dado')
plt.ylabel('Frecuencia Relativa')
plt.title('Histograma de 1000 Lanzamientos de un Dado Truncado')

# Graficar la distribución teórica sobre el histograma
plt.plot(valores, probabilidades, 'ro-', label='Distribución Teórica')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Comentario sobre la aproximación a la distribución teórica
print("Para que los resultados se aproximen más a la distribución teórica, sería necesario incrementar el número de simulaciones. Según la Ley de los Grandes Números, a medida que aumentamos el número de lanzamientos, la distribución empírica se aproximará a la distribución teórica.")

# Simulación del inciso b)
# Simular 1000 veces la suma de 5 lanzamientos de dados truncados
suma_lanzamientos = [np.sum(np.random.choice(valores, size=5, p=probabilidades)) for _ in range(1000)]

# Graficar el histograma de las sumas de los lanzamientos
plt.figure(figsize=(10, 5))
plt.hist(suma_lanzamientos, bins=range(5, 31), density=True, rwidth=0.8, color='lightgreen', edgecolor='black', align='mid')
plt.xlabel('Suma de 5 Lanzamientos')
plt.ylabel('Frecuencia Relativa')
plt.title('Histograma de la Suma de 5 Lanzamientos de un Dado Truncado')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Explicación sobre el Teorema del Límite Central
print("Según el Teorema del Límite Central, la suma de variables aleatorias independientes tiende a distribuirse de manera aproximadamente normal a medida que aumenta el número de variables sumadas. Por lo tanto, en este caso, la distribución de las sumas debería aproximarse a una distribución normal.")
```

b) Simule 1000 veces la suma de 5 dados trucados (con las mismas probabilidades
mencionadas). Grafique el histograma de las sumas. ¿Cómo se distribuyen las sumas?
Explique el comportamiento considerando el Teorema del Límite Central. (1.0 pts.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del dado truncado
# Dados truncados con probabilidades diferentes para los valores {1, 2, 3, 4, 5, 6}
valores = np.array([1, 2, 3, 4, 5, 6])
probabilidades = np.array([0.1, 0.2, 0.3, 0.2, 0.1, 0.1])  # Probabilidades asignadas

# Número de simulaciones
n_simulaciones = 1000
n_dados = 5  # Número de dados

# Simular las sumas de los lanzamientos de 5 dados truncados
suma_dados = np.sum(
    np.random.choice(valores, size=(n_simulaciones, n_dados), p=probabilidades), axis=1
)

# Crear histograma
plt.figure(figsize=(10, 6))
plt.hist(suma_dados, bins=30, density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title("Distribución de las Sumas de 5 Dados Truncados", fontsize=14)
plt.xlabel("Suma", fontsize=12)
plt.ylabel("Frecuencia Relativa", fontsize=12)
plt.grid(True)

# Mostrar la gráfica
plt.show()

# Calcular la media y desviación estándar de las sumas
media_sumas = np.mean(suma_dados)
desviacion_sumas = np.std(suma_dados)

print(f"Media de las sumas: {media_sumas}")
print(f"Desviación estándar de las sumas: {desviacion_sumas}")
```
