```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Cargar los datos desde el archivo 's1_s2.txt' omitiendo la primera línea
data = np.loadtxt('s1_s2.txt', skiprows=1)
s1 = data[:, 0]  # Primera columna
s2 = data[:, 1]  # Segunda columna
```

b) (Tarea asíncrona) Indique si, para s1 y s2, la autocorrelación es igual a la
autocovarianza. Realice las gráficas que considere pertinentes para fundamentar su
respuesta. (1.0 pts.)

```python
# Función a utilizar:
def CalcularAutocovarianza(Arreglo):
    N = len(Arreglo)
    media = np.mean(Arreglo)
    autocovar = np.correlate(Arreglo - media, Arreglo - media, mode='full') / N
    return autocovar
# Cálculos
autocovarx1 = CalcularAutocovarianza(s1)
autocorrx1 = (np.correlate(s1,s1,mode='full'))/(len(s1))
autocovarx2 = CalcularAutocovarianza(s2)
autocorrx2 = (np.correlate(s2,s2,mode='full'))/(len(s2))
# Gráficas
tau = np.arange(len(autocorrx1))-len(autocorrx1)/2
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.plot(tau, autocorrx1, label='Autocorrelación x1', color='#5b95bd')
plt.title('Autocorrelación de x1')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocorrelación')
plt.legend()
plt.grid()
plt.subplot(2, 2, 2)
plt.plot(tau, autocovarx1, label='Autocovarianza x1', color='#5b95ff')
plt.title('Autocovarianza de x1')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocovarianza')
plt.grid()
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(tau, autocorrx2, label='Autocorrelación x2', color='#5bbda9')
plt.title('Autocorrelación de x2')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocorrelación')
plt.grid()
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(tau, autocovarx2, label='Autocovarianza x2',color='#5bbdcf')
plt.title('Autocovarianza de x2')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocovarianza')
plt.grid()
plt.legend()
plt.tight_layout()
# Impresión de algunos valores para comparar ambos cálculos:
print("-------------------------------------------------------------------------------------------")
for tauAux in range(5):
    print(f"Ronda N°{tauAux+1} de valores:")
    print(f"Autocorrelación x1 = {autocorrx1[999+tauAux]}, Autocovarianza x1 = {autocovarx1[999+tauAux]}")
    print(f"Autocorrelación x2 = {autocorrx2[999+tauAux]}, Autocovarianza x2 = {autocovarx2[999+tauAux]}")
    print("-------------------------------------------------------------------------------------------")
```
