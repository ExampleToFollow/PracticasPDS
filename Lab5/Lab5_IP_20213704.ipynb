El ruido blanco de tipo gaussiano es considerado así debido a su natuleza de encontrarse inmerso en todas las frecuencias , tener una media de cero y estar altamente autocorrelacionado . Esto ultimo se interpreta como que una muestra en el tiempo solo es similar a si misma y aninguna otra muestra. Al realizar las gráficas preliminares de las estadísticas del sistema podemos considerar que las medias de las señales son cercanas a cero y las graficas de los histogramas se asemejan a una campana de Gauss por lo que podemos sospechar q ambas señales serían candidatos a ser gaussianas con media cercana a cero pero no cumplen con el ultimo criterio de la autocovarianza alta en cero y casi nula en las otras componentes como se ve en la siguiente pregunta. Donde se terminará concluyendo que la señal s1 es la q más se asemeja a una producida por ruido blanco gaussiano.

b) (Tarea asíncrona) Indique si, para s1 y s2, la autocorrelación es igual a la
autocovarianza. Realice las gráficas que considere pertinentes para fundamentar su
respuesta. (1.0 pts.)

```python
# Función a utilizar:
def CalcularAutocovarianza(signal):
    return np.correlate(signal - np.mean(signal), signal - np.mean(signal), mode='full') / len(signal)
# Cálculos
autocovarx1 = CalcularAutocovarianza(s1)
autocorrx1 = (np.correlate(s1,s1,mode='full'))/(len(s1))
autocovarx2 = CalcularAutocovarianza(s2)
autocorrx2 = (np.correlate(s2,s2,mode='full'))/(len(s2))
# Gráficas
rango = np.arange(len(autocorrx1))-len(autocorrx1)/2
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.plot(rango, autocorrx1, label='Autocorrelación s1', color='#5b95bd')
plt.title('Autocorrelación de s1')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocorrelación')
plt.legend()
plt.grid()
plt.subplot(2, 2, 2)
plt.plot(rango, autocovarx1, label='Autocovarianza s1', color='#5b95ff')
plt.title('Autocovarianza de s1')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocovarianza')
plt.grid()
plt.legend()
plt.subplot(2, 2, 3)
plt.plot(rango, autocorrx2, label='Autocorrelación s2', color='#5bbda9')
plt.title('Autocorrelación de s2')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocorrelación')
plt.grid()
plt.legend()
plt.subplot(2, 2, 4)
plt.plot(rango, autocovarx2, label='Autocovarianza s2',color='#5bbdcf')
plt.title('Autocovarianza de s2')
plt.xlabel(r'$\tau$')
plt.ylabel('Autocovarianza')
plt.grid()
plt.legend()
plt.tight_layout()

print(f'>> Media de s1: {np.mean(s1)}')
print(f'>> Media de s2: {np.mean(s2)}')
```

Como se puede apreciar en los gráficos obtenidos la autocorrelación y la autocovarianza de cada señal son iguales , esto se debe a que las medias de ambas señales son bastante cercanas a cero de tal forma q la definición de autocovarianza y autocorrelación se volvieron similares Autocorrelación =  E{X[n] ^ 2 } Autocovarianza = E{ (X[n] - mu )^2} con mu->0  entonces => Autocovarianza = Autocorrelación esto se puede apreciar de manera gráfica pero si vemos muestra por muestra no lo son con tanta precisión. Además cabe resaltar que al obtener estas gráficas podemos dicernir sobre cual de ellas es ruido blanco gaussiano la cual es la s1 al estar altamente autocorrelacionada en cero.
