```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Cargar los datos desde el archivo 's1_s2.txt' omitiendo la primera línea
data = np.loadtxt('s1_s2.txt', skiprows=1)
s1 = data[:, 0]  # Primera columna
s2 = data[:, 1]  # Segunda columna
```

```python
N = len(s1)
x = s1
n = np.arange(N)
xrange = [-3,3]
Nb = 13
dx = (xrange[1] - xrange[0]) / (Nb-1)
bins = xrange[0]-dx/2 + np.arange(Nb+1)*dx
cbin = xrange[0] + np.arange(Nb)*dx
plt.figure()
plt.plot(n,x,'-o')
plt.grid('on')
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.ylim((bins[0]*1.1,bins[-1]*1.1))
plt.figure()
plt.hist(x,bins=bins,orientation='horizontal',rwidth=0.8,align='mid',ec='k')
plt.grid('on')

plt.figure()
plt.hist(x,bins=bins,rwidth=0.8,align='mid',ec='k')
plt.grid('on')

plt.figure()

h,_,_ = plt.hist(x,bins=bins,rwidth=0.8,align='mid',cumulative=True, ec='k')
plt.plot(cbin,h,'ro-')
plt.grid('on')
```
